Implement ABC wit $\epsilon=0.3$ with cont. data.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sc

from scipy.stats import beta
np.random.seed(0)
post_abc=list()
y_gen=list()
theta_gen=list()
post_y=list()
for i in range(20000):
  p = np.random.normal(0,1)
  s = np.random.normal(p, 1)
  y_gen.append(s)
  theta_gen.append(p)  
  if 0.7<s<1.3:
    post_abc.append(p)
    post_y.append(s)
    
```

Plot the accepted simulations. 

```python
post=zip(post_abc,post_y)
post=list(post)
plt.scatter(y_gen,theta_gen,alpha=0.5,s=10)
plt.scatter( post_y,post_abc,alpha=0.5,s=10,c="red")
plt.axvline(x=0.7,c='red')
plt.axvline(x=1.3,c='red')
plt.xlabel('y')
_=plt.ylabel('$\\theta$')
plt.savefig('C:/Users/Jost/Desktop/Master/WS 19_20/Research Seminar in Statistics Bayesian computation/Presentation/Figures/scatter_cont_eps.png')
```

Define normal likelihood.

```python
import math
def norm_lik(mu,sigma,obs):
    return((1/sigma*(math.sqrt(2*np.pi)))*np.exp(-(obs-mu)**2/2*sigma**2))
    
```

Calculate actual posterior and plot it with sample.

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
x = np.linspace(norm.ppf(0.01),norm.ppf(0.99), 100)

fig, ax1 = plt.subplots()

ax1.set_xlabel(r'$\theta$')
ax1.set_ylabel('Samples', color='blue')
ax1.hist(post_abc,bins=50)
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'red'
ax2.set_ylabel('Density', color=color)
ax2.set_yticklabels([])

ax2.plot(x, norm_lik(x,1,1)*norm.pdf(x),c="red")
ax2.tick_params(axis='y', labelcolor='red')
_=ax2.set_ylim([0, 0.9])
plt.savefig('C:/Users/Jost/Desktop/Master/WS 19_20/Research Seminar in Statistics Bayesian computation/Presentation/Figures/cont_dens_hist.png')
```

```python
import scipy.stats as stats
import math
mu = 1
variance = 0.1
sigma = math.sqrt(variance)
def norm_gaus(x):
    return stats.norm.pdf(x, 1, 0.2)/(stats.norm.pdf(1.95, 1, 0.2))
```

Introduce normal Kernel and accept with a pobability prop. to distance. 

```python
np.random.seed(0)
post_abc4=list()
y_gen4=list()
theta_gen4=list()
post_y4=list()
for i in range(20000):
  p = np.random.normal(0,1)
  s = np.random.normal(p, 1)
  u = np.random.uniform(0,1)  
  y_gen4.append(s)
  theta_gen4.append(p)  
  if norm_gaus(s)>u:
    post_abc4.append(p)
    post_y4.append(s)
```

Plot the kernel accepted simulations.

```python
mu = 1
variance = 0.2
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma)-4,label="Gaussian Kernel")

post4=zip(post_abc4,post_y4)
post4=list(post4)
plt.scatter(y_gen4,theta_gen4,alpha=0.5,s=10)
plt.scatter( post_y4,post_abc4,alpha=0.5,s=10,c="red",label="accepted obs.")
plt.xlabel('y')
plt.legend()
_=plt.ylabel('$\\theta$')
plt.savefig('C:/Users/Jost/Desktop/Master/WS 19_20/Research Seminar in Statistics Bayesian computation/Presentation/Figures/cont_scatter_kernel.png')
```

Compare both methods. 

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
x = np.linspace(norm.ppf(0.01),norm.ppf(0.99), 100)

fig, ax1 = plt.subplots()

ax1.set_xlabel(r'$\theta$')
ax1.set_ylabel('', color='blue')
plt.hist(post_abc4,bins=50,density=True,color="C0",alpha=0.6,label="Kernel")
plt.hist(post_abc2,bins=50,density=True,color="C0",label="Huge $\epsilon$")
plt.legend(loc="upper left")


ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
ax2.set_yticklabels([])

color = 'red'
ax2.set_ylabel('', color=color)

ax2.plot(x, norm_lik(x,1,1)*norm.pdf(x),c="red",label="True density")
ax2.tick_params(axis='y', labelcolor='red')
plt.legend(loc="upper right")
_=ax2.set_ylim([0, 0.9])
plt.savefig('C:/Users/Jost/Desktop/Master/WS 19_20/Research Seminar in Statistics Bayesian computation/Presentation/Figures/cont_dens_hist_kernel_eps.png')
```
