```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import beta
np.random.seed(0)
post_abc=list()
y_gen=list()
theta_gen=list()
post_y=list()
for i in range(100000):
  p = np.random.beta(1,1)
  s = np.random.binomial(10, p)
  y_gen.append(s)
  theta_gen.append(p)  
  if s==7:
    post_abc.append(p)
    post_y.append(s)
    
```

```python
plt.scatter(y_gen,theta_gen,alpha=0.5,s=10)
```

```python
post=zip(post_abc,post_y)
post=list(post)
plt.scatter(y_gen,theta_gen,alpha=0.5,s=10)
plt.scatter([7]*len(post),post_abc,alpha=0.5,s=10,c="red")
    
```

```python
plt.hist(post_abc,bins=50)
axes = plt.gca()
axes.set_xlim([0,1])
```

```python
x = np.linspace(0, 1, 1002)[1:-1]
dist = beta(a=8,b=4)
plt.plot(x,dist.pdf(x),c="red")
```
