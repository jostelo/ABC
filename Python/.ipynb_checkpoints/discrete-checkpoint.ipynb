Implement simple ABC Algorithm

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import beta
np.random.seed(0)
post_abc=list()
y_gen=list()
theta_gen=list()
post_y=list()
for i in range(20000):
  p = np.random.beta(1,1)
  s = np.random.binomial(10, p)
  y_gen.append(s)
  theta_gen.append(p)  
  if s==7:
    post_abc.append(p)
    post_y.append(s)
    
```

Plot the data

```python
plt.scatter(y_gen,theta_gen,alpha=0.5,s=10)
plt.xlabel('y')
_=plt.ylabel('$\\theta$')
plt.savefig('C:/Users/Jost/Desktop/Master/WS 19_20/Research Seminar in Statistics Bayesian computation/Presentation/Figures/discrete.png')
```

```python
post=zip(post_abc,post_y)
post=list(post)
plt.scatter(y_gen,theta_gen,alpha=0.5,s=10)
plt.scatter([7]*len(post),post_abc,alpha=0.5,s=10,c="red")
plt.xlabel('y')
_=plt.ylabel('$\\theta$')
plt.savefig('C:/Users/Jost/Desktop/Master/WS 19_20/Research Seminar in Statistics Bayesian computation/Presentation/Figures/discrete2.png')    
```

Plot histogram

```python
plt.hist(post_abc,bins=50)
axes = plt.gca()
axes.set_xlim([0,1])
plt.ylabel('samples')
_=plt.xlabel('$\\theta$')
```

Calculate actual posterior and plot it also. 

```python
x = np.linspace(0, 1, 1002)[1:-1]
dist = beta(a=8,b=4)
t = np.arange(0.01, 10.0, 0.01)
data1 = np.exp(t)
data2 = np.sin(2 * np.pi * t)

fig, ax1 = plt.subplots()


ax1.set_xlabel(r'$\theta$')
ax1.set_ylabel('Samples', color='blue')
ax1.hist(post_abc,bins=50)
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'red'
ax2.set_ylabel('Density', color=color)  
ax2.plot(x,dist.pdf(x),c="red")
ax2.tick_params(axis='y', labelcolor='red')
ax2.set_ylim([0, 3.5])

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.savefig('C:/Users/Jost/Desktop/Master/WS 19_20/Research Seminar in Statistics Bayesian computation/Presentation/Figures/histogram_dens.png')
```
